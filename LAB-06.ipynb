```python
# Lab 9: Visualize Relationships Between Variables Using Seaborn Pair Plots
# Scenario 6: Wine Quality Assessment

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (make sure the file path is correct)
df = pd.read_csv("/content/winequality-red.csv")

# Select important numerical features
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'alcohol', 'quality']
wine_data = df[features]

# Set Seaborn style
sns.set(style="whitegrid")

# Create a pair plot with 'quality' as hue
sns.pairplot(wine_data,
             hue="quality",
             palette="coolwarm",
             diag_kind="kde")

# Add title
plt.suptitle("Pairwise Relationships Among Wine Chemical Properties", y=1.02)
plt.show()
```

```python
# Lab 9: Visualize Relationships Between Variables Using Seaborn Pair Plots
# Scenario 5: Student Performance Visualization

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("/content/student_performance.csv")

# Step 2: Check data
print("Dataset preview:")
print(df.head(), "\n")

# Step 3: Optional  Normalize numerical data (scaling)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[['StudyTime', 'Attendance', 'ExamScore']] = scaler.fit_transform(
    df_scaled[['StudyTime', 'Attendance', 'ExamScore']]
)

# Step 4: Visualization using Seaborn Pair Plot
sns.set(style="whitegrid")
pair_plot = sns.pairplot(
    df_scaled,
    vars=['StudyTime', 'Attendance', 'ExamScore'],
    hue='Gender',
    palette='coolwarm',
    diag_kind='kde'
)

# Step 5: Title and layout
pair_plot.fig.suptitle("Student Performance: Study Time, Attendance, and Exam Scores", y=1.02)
plt.show()
```
